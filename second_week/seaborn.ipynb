```python
# KDE plots


mean=[0,0]
cov=[[1,0.4],[0.4,1]]
x,y=np.random.multivariate_normal(mean,cov,size=40).T
# print(x,y)
sns.kdeplot(y,fill=True,color='green')
plt.show()


```

```python
sns.kdeplot(x=x,y=y,fill=False,cmap='coolwarm',n_levels=35)
plt.show()
sns.kdeplot(y=x,fill=False,bw_method=3)
plt.show()
```

```python
# joint plot
sns.jointplot(x='total_bill',y='tip',data=tips,ratio=2,size=6)
# kind can be reg,kde,
```

```python

from scipy.stats import spearmanr

# create jointplot
g = sns.jointplot(x='total_bill', y='tip', data=tips)

# calculate spearman correlation
corr, _ = spearmanr(tips['total_bill'], tips['tip'])

# add text annotation
g.ax_joint.text(
    0.05, 0.95,
    f"Spearman r = {corr:.2f}",
    transform=g.ax_joint.transAxes
)

plt.show()
```
