```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ages_x=np.arange(23,36)
dev_y=np.floor(np.random.random(13)*10000)
python_dev_y=np.floor(np.random.random(13)*10000)
js_dev_y=np.floor(np.random.random(13)*10000)
x_index=np.arange(len(ages_x))
width=0.25

plt.barh(x_index-width,dev_y,color='k',label='developer',height=width)
plt.barh(x_index,python_dev_y,color='b',label='python',height=width)
plt.barh(x_index+width,js_dev_y,color='r',label='js',height=width)
plt.ylabel("ages")
plt.xlabel("salary")
plt.title("age v/s salary")
plt.yticks(ticks=x_index,labels=ages_x)
plt.legend()
plt.tight_layout()
plt.grid(True)
# plt.xscale(1000000)
# plt.ylim(50)
plt.show()


```

```python
ages_x=np.arange(23,36)
dev_y=np.floor(np.random.random(13)*10000)
python_dev_y=np.floor(np.random.random(13)*10000)

plt.plot(ages_x,dev_y,label='dev',color='red')
plt.plot(ages_x,python_dev_y,label='python',linestyle='--')

plt.fill_between(ages_x,python_dev_y,np.median(python_dev_y),alpha=0.25,where=(python_dev_y>np.median(python_dev_y)),interpolate=True)
plt.fill_between(ages_x,python_dev_y,np.median(python_dev_y),alpha=0.25,where=(python_dev_y<=np.median(python_dev_y)),interpolate=True)
plt.ylabel("salary")
plt.xlabel("ages")
plt.title("age v/s salary")
# plt.xticks(ticks=x_index,labels=ages_x)
plt.legend()
plt.tight_layout()
plt.grid(True)
# plt.xscale(1000000)
# plt.ylim(50)
plt.show()
```

```python
ages=[18,19,21,25,26,26,30,32,38,45,55]
bins=[10,20,30,40,50,60]

plt.hist(ages,bins=bins,edgecolor='black',log=True)
median_age=np.median(ages)
color='#fc4f30'
plt.axvline(median_age,color=color,label='age median',linewidth=2)
plt.title("ages")
plt.xlabel("age axis")
plt.ylabel('total')
plt.tight_layout()
plt.legend()
plt.grid(True)
plt.show()
```

```python
# time series data
from datetime import datetime,timedelta
from matplotlib import dates as mpl_dates
dates=[
    datetime(2019,5,24),
    datetime(2019,3,22),
    datetime(2019,5,30),
    datetime(2019,8,2),
    datetime(2019,12,1),
    datetime(2019,10,4),
    datetime(2019,3,16),
    datetime(2019,5,10),
    datetime(2019,9,25),
    datetime(2019,7,30),
    datetime(2019,6,22),
]

y=np.floor(np.random.rand(len(dates))*10)
plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate()
date_format=mpl_dates.DateFormatter('%b, %d ,%Y')
plt.gca().xaxis.set_major_formatter(date_format)
# plt.xticks()
plt.tight_layout()
plt.show()
```

```python
# !pip install IPython
import numpy as np
import matplotlib.pyplot as plt
from itertools import count
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

x_vals = []
y_vals = []
index = count()

fig, ax = plt.subplots()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(np.random.randint(0, 5))

    ax.clear()
    ax.plot(x_vals, y_vals)
    ax.grid(True)
    ax.set_title("Real-Time Data Plot")

ani = FuncAnimation(
    fig,
    animate,
    interval=1000,
    cache_frame_data=False
)

HTML(ani.to_jshtml())
```
